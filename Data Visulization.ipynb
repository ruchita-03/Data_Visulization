```python
import pandas as pd

# 1. CSV file importing
df_csv = pd.read_csv('etrm_trades.csv')
print("CSV DataFrame:")
display(df_csv)
```

```python
# 2. Converting json file 
df_json=pd.read_json('etrm_trades.json')
print("JSON DataFrame: ")
display(df_json)

```

```python
#3. converting the excel file into dataframe
df_excel=pd.read_excel('etrm_trades.xlsx',sheet_name=0)
print(" Excel DataFrame: ")
display(df_excel)
```

```python
#4. Txt file converting into DataFrame
df_txt=pd.read_csv('etrm_trades.txt',sep='\t')
print("Txt DataFrame is  :")
display(df_txt)
```

```python
# 5. HTML
df_html=pd.read_html('etrm_trades.html')[0]
print("HTML DataFrame:")
display(df_html)
```

```python
from lxml import etree

# converting the xml file into DataFrame
tree = etree.parse('etrm_trades.xml')
root = tree.getroot()
data = []
for trade in root.findall('Trade'):
    trade_data = {
        'TradeID': int(trade.find('TradeID').text),
        'Commodity': trade.find('Commodity').text,
        'TradeType': trade.find('TradeType').text,
        'Volume': float(trade.find('Volume').text),
        'Price': float(trade.find('Price').text),
        'Currency': trade.find('Currency').text,
        'DeliveryStart': trade.find('DeliveryStart').text,
        'DeliveryEnd': trade.find('DeliveryEnd').text,
        'Periodicity': trade.find('Periodicity').text
    }
    data.append(trade_data)

df_xml = pd.DataFrame(data)
print("XML DataFrame:")
display(df_xml)
```

```python

# Convert both columns to datetime because they are imported as string
df_csv['DeliveryStart'] = pd.to_datetime(df_csv['DeliveryStart'], errors='coerce')
df_csv['DeliveryEnd'] = pd.to_datetime(df_csv['DeliveryEnd'], errors='coerce')


print(df_csv.dtypes)

```

```python
#Average Price per Commodity
# Group by Commodity and calculate average price
avg_price_per_commodity = df_csv.groupby('Commodity')['Price'].mean().sort_values(ascending=False)

print("Average Price per Commodity:")
print(avg_price_per_commodity)
```

```python
#Distribution of Trades by Currency
import matplotlib.pyplot as plt

# Count trades by Currency
currency_distribution = df_csv['Currency'].value_counts()

print("Distribution of Trades by Currency:")
print(currency_distribution)

currency_distribution.plot(kind='bar', figsize=(6,4), title="Trades Distribution by Currency")
plt.xlabel("Currency")
plt.ylabel("Number of Trades")
plt.show()
```

```python
#Trade Periodicity Breakdown
# Ensure DeliveryStart is datetime
df_csv['DeliveryStart'] = pd.to_datetime(df_csv['DeliveryStart'], errors='coerce')

# Daily trades
daily_trades = df_csv.groupby(df_csv['DeliveryStart'].dt.date).size()

# Weekly trades
weekly_trades = df_csv.groupby(df_csv['DeliveryStart'].dt.to_period('W')).size()

# Monthly trades
monthly_trades = df_csv.groupby(df_csv['DeliveryStart'].dt.to_period('M')).size()

print("Daily Trades:\n", daily_trades.head())
print("\nWeekly Trades:\n", weekly_trades.head())
print("\nMonthly Trades:\n", monthly_trades.head())

```

```python
#Histogram: Distribution of Notional Values

import seaborn as sns

plt.figure(figsize=(10,6))
sns.histplot(df_csv['notional'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of Notional Values', fontsize=16)
plt.xlabel('Notional Value')
plt.ylabel('Frequency')
plt.show()

```

```python
#Heatmap: Trade Counts by Commodity and Trade Type
import matplotlib.pyplot as plt
import seaborn as sns

#  Commodity vs Trade Type counts
trade_counts = df_csv.pivot_table(
    index='commodity',
    columns='tradetype',
    values='volume',   # lowercase
    aggfunc='count',
    fill_value=0
)

plt.figure(figsize=(12,6))
sns.heatmap(trade_counts, annot=True, fmt='d', cmap='YlGnBu')
plt.title('Trade Counts: Commodity vs Trade Type', fontsize=16)
plt.xlabel('Trade Type')
plt.ylabel('Commodity')
plt.show()

```

```python
#Scatter Plot: Price vs Volume by Commodity
plt.figure(figsize=(10,6))
sns.scatterplot(
    data=df_csv,
    x='volume',
    y='price',
    hue='commodity',
    palette='tab10'
)
plt.title('Price vs Volume by Commodity', fontsize=16)
plt.xlabel('Volume')
plt.ylabel('Price')
plt.legend(title='Commodity', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
