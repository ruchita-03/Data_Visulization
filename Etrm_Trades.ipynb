```python
import pandas as pd
from lxml import etree

# Load CSV 
df_csv = pd.read_csv('etrm_trades.csv')

# Load JSON 
df_json = pd.read_json('etrm_trades.json')

# Load Excel
df_excel = pd.read_excel('etrm_trades.xlsx', sheet_name=0)

#  Load TXT 
df_txt = pd.read_csv('etrm_trades.txt', sep='\t')

# Load HTML 
df_html = pd.read_html('etrm_trades.html')[0]

# Load XML
tree = etree.parse('etrm_trades.xml')
root = tree.getroot()
data = []

for trade in root.findall('Trade'):
    trade_data = {
        'TradeID': int(trade.find('TradeID').text),
        'Commodity': trade.find('Commodity').text,
        'TradeType': trade.find('TradeType').text,
        'Volume': float(trade.find('Volume').text),
        'Price': float(trade.find('Price').text),
        'Currency': trade.find('Currency').text,
        'DeliveryStart': trade.find('DeliveryStart').text,
        'DeliveryEnd': trade.find('DeliveryEnd').text,
        'Periodicity': trade.find('Periodicity').text
    }
    data.append(trade_data)

df_xml = pd.DataFrame(data)


def clean_columns(df):
    df.columns = df.columns.str.strip().str.lower().str.replace(" ", "")
    return df

df_csv = clean_columns(df_csv)
df_json = clean_columns(df_json)
df_excel = clean_columns(df_excel)
df_txt = clean_columns(df_txt)
df_html = clean_columns(df_html)
df_xml = clean_columns(df_xml)


expected_columns = df_csv.columns


datasets = [df_csv, df_json, df_excel, df_txt, df_html, df_xml]
for i in range(len(datasets)):
    datasets[i] = datasets[i].reindex(columns=expected_columns)

#  Merge all into a single DataFrame 
all_trades = pd.concat(datasets, ignore_index=True)

#  Display the full combined DataFrame 
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

display(all_trades)


all_trades.to_csv("all_trades.csv", index=False)

```

```python
# Drop all rows with any null values
all_trades_clean = all_trades.dropna()


all_trades_clean.reset_index(drop=True, inplace=True)

#  the cleaned DataFrame
display(all_trades_clean)


all_trades_clean.to_csv("all_trades_clean.csv", index=False)

```

```python
import pandas as pd


all_trades_clean = all_trades_clean.copy()

# Convert DeliveryStart and DeliveryEnd to datetime because they are imported as string 
all_trades_clean.loc[:, 'deliverystart'] = pd.to_datetime(all_trades_clean['deliverystart'], errors='coerce')
all_trades_clean.loc[:, 'deliveryend'] = pd.to_datetime(all_trades_clean['deliveryend'], errors='coerce')


all_trades_clean.reset_index(drop=True, inplace=True)


print(all_trades_clean)


all_trades_clean.to_csv("all_trades_clean_datetime.csv", index=False)
```

```python

# a.Average price per commodity 
avg_price_per_commodity = all_trades_clean.groupby('commodity')['price'].mean().reset_index()
avg_price_per_commodity.rename(columns={'price': 'average_price'}, inplace=True)
print("Average Price per Commodity:")
print(avg_price_per_commodity)
print("\n" + "="*50 + "\n")

# b. Distribution of trades by currency
trades_by_currency = all_trades_clean['currency'].value_counts().reset_index()
trades_by_currency.columns = ['currency', 'trade_count']
print("Distribution of Trades by Currency:")
print(trades_by_currency)
print("\n" + "="*50 + "\n")

# c. Trade periodicity breakdown 
trade_periodicity = all_trades_clean['periodicity'].value_counts().reset_index()
trade_periodicity.columns = ['periodicity', 'trade_count']
print("Trade Periodicity Breakdown:")
print(trade_periodicity)
print("\n" + "="*50 + "\n")
```

```python
#Top 5 Commodities by Volume
top_commodities_volume = all_trades_clean.groupby('commodity')['volume'].sum().reset_index()
top_commodities_volume = top_commodities_volume.sort_values(by='volume', ascending=False)
print("Top 5 Commodities by Total Volume:")
print(top_commodities_volume.head(5))
print("\n" + "="*50 + "\n")

#Total Trade Value per Commodity

all_trades_clean['trade_value'] = all_trades_clean['volume'] * all_trades_clean['price']

trade_value_per_commodity = all_trades_clean.groupby('commodity')['trade_value'].sum().reset_index()
trade_value_per_commodity = trade_value_per_commodity.sort_values(by='trade_value', ascending=False)
print("Total Trade Value per Commodity:")
print(trade_value_per_commodity)
print("\n" + "="*50 + "\n")

#Trades Over Time 
# Convert DeliveryStart to datetime if not already done
all_trades_clean['deliverystart'] = pd.to_datetime(all_trades_clean['deliverystart'], errors='coerce')

# Count number of trades per day
trades_per_day = all_trades_clean.groupby(all_trades_clean['deliverystart'].dt.date).size().reset_index(name='trade_count')
print("Number of Trades per Day:")
print(trades_per_day.head(10))  # shows first 10 days
print("\n" + "="*50 + "\n")

#Average Price per Trade Type
avg_price_per_tradetype = all_trades_clean.groupby('tradetype')['price'].mean().reset_index()
avg_price_per_tradetype.rename(columns={'price':'average_price'}, inplace=True)
print("Average Price per Trade Type:")
print(avg_price_per_tradetype)
print("\n" + "="*50 + "\n")



```

```python
import pandas as pd
from lxml import etree

# Load CSV 
df_csv = pd.read_csv('etrm_trades.csv')

# Load JSON 
df_json = pd.read_json('etrm_trades.json')

# Load Excel
df_excel = pd.read_excel('etrm_trades.xlsx', sheet_name=0)

#  Load TXT 
df_txt = pd.read_csv('etrm_trades.txt', sep='\t')

# Load HTML 
df_html = pd.read_html('etrm_trades.html')[0]

# Load XML
tree = etree.parse('etrm_trades.xml')
root = tree.getroot()
data = []

for trade in root.findall('Trade'):
    trade_data = {
        'TradeID': int(trade.find('TradeID').text),
        'Commodity': trade.find('Commodity').text,
        'TradeType': trade.find('TradeType').text,
        'Volume': float(trade.find('Volume').text),
        'Price': float(trade.find('Price').text),
        'Currency': trade.find('Currency').text,
        'DeliveryStart': trade.find('DeliveryStart').text,
        'DeliveryEnd': trade.find('DeliveryEnd').text,
        'Periodicity': trade.find('Periodicity').text
    }
    data.append(trade_data)

df_xml = pd.DataFrame(data)


def clean_columns(df):
    df.columns = df.columns.str.strip().str.lower().str.replace(" ", "")
    return df

df_csv = clean_columns(df_csv)
df_json = clean_columns(df_json)
df_excel = clean_columns(df_excel)
df_txt = clean_columns(df_txt)
df_html = clean_columns(df_html)
df_xml = clean_columns(df_xml)


expected_columns = df_csv.columns


datasets = [df_csv, df_json, df_excel, df_txt, df_html, df_xml]
for i in range(len(datasets)):
    datasets[i] = datasets[i].reindex(columns=expected_columns)

#  Merge all into a single DataFrame 
all_trades = pd.concat(datasets, ignore_index=True)

#  Display the full combined DataFrame 
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

display(all_trades)


all_trades.to_csv("all_trades.csv", index=False)


```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Set Seaborn style
sns.set(style="whitegrid", font_scale=1.1)
plt.rcParams["figure.figsize"] = (18, 24)

# Preparing data
volume_by_commodity = all_trades_clean.groupby('commodity')['volume'].sum().sort_values(ascending=False)
trades_by_currency = all_trades_clean['currency'].value_counts()
avg_price_by_date = all_trades_clean.groupby('deliverystart')['price'].mean().reset_index()
all_trades_clean['notional'] = all_trades_clean['volume'] * all_trades_clean['price']
commodity_tradetype_counts = all_trades_clean.pivot_table(
    index='commodity', columns='tradetype', values='tradeid', aggfunc='count', fill_value=0
)

# Create dashboard
fig, axes = plt.subplots(3, 2, figsize=(18, 24))

colors = sns.color_palette("Spectral", len(volume_by_commodity))

#1.Bar chart: Volume by trader.  
volume_df = volume_by_commodity.reset_index().rename(columns={'volume':'total_volume'})

sns.barplot(
    data=volume_df,
    x='commodity',
    y='total_volume',
    hue='commodity',  
    dodge=False,
    palette=colors,
    legend=False,     
    ax=axes[0,0]
)

axes[0,0].set_title("Total Volume by Commodity", fontsize=16, weight='bold')
axes[0,0].set_xlabel("Commodity")
axes[0,0].set_ylabel("Total Volume")
axes[0,0].tick_params(axis='x', rotation=45)


# 2. Pie chart: Trades by Currency 
axes[0,1].pie(trades_by_currency.values, labels=trades_by_currency.index, autopct='%1.1f%%', startangle=140,
              colors=sns.color_palette("Set3", len(trades_by_currency)), textprops={'fontsize':12, 'weight':'bold'})
axes[0,1].set_title("Distribution of Trades by Currency", fontsize=16, weight='bold')

# 3. Line chart: Average Price Trend
sns.lineplot(data=avg_price_by_date, x='deliverystart', y='price', marker='o', color='mediumvioletred', linewidth=2.5, markersize=8, ax=axes[1,0])
axes[1,0].set_title("Average Price Trend by Delivery Start Date", fontsize=16, weight='bold')
axes[1,0].set_xlabel("Delivery Start Date")
axes[1,0].set_ylabel("Average Price")
axes[1,0].tick_params(axis='x', rotation=45)

# 4. Histogram: Distribution of Notional Values 
sns.histplot(all_trades_clean['notional'], bins=30, kde=True, color='darkorange', edgecolor='black', alpha=0.7, ax=axes[1,1])
axes[1,1].set_title("Distribution of Notional Values", fontsize=16, weight='bold')
axes[1,1].set_xlabel("Notional Value")
axes[1,1].set_ylabel("Frequency")

# 5. Heatmap: Commodity vs TradeType Trade Counts 
sns.heatmap(commodity_tradetype_counts, annot=True, fmt="d", cmap="RdYlGn_r", cbar_kws={'label':'Trade Counts'}, linewidths=0.5, linecolor='gray', ax=axes[2,0])
axes[2,0].set_title("Commodity vs Trade Type Trade Counts", fontsize=16, weight='bold')
axes[2,0].set_xlabel("Trade Type")
axes[2,0].set_ylabel("Commodity")


fig.delaxes(axes[2,1])

plt.tight_layout()
plt.show()

# Saving the  dashboard 
fig.savefig("trade_dashboard_colorful.png", dpi=300)
```
